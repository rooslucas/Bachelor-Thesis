```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#plt.figure(figsize=(10, 10))
plt.rcParams.update({'font.family': 'Times'})
plt.rcParams.update({'font.size': 12})

data_file_path = '/Users/roos/Data/bootstrap/bootstrapped_data.csv'
output = pd.read_csv(data_file_path)
```

```python
n, bins, patches2= plt.hist(x=output['diffa_o1-b'], bins='auto', color='lightcoral', rwidth=0.5)

plt.xlabel('Acc difference')
plt.ylabel('Frequency')
#plt.title('Accuracy Difference between Model 1 and Baseline')
maxfreq = n.max()

# confidence intervals
alpha = 0.95
p = (1.0-alpha) * 100
lower =  np.percentile(output['diffa_o1-b'], p)
p = (alpha+((1.0-alpha)/2.0)) * 100
upper =  max(output['diffa_o1-b']) #np.percentile(output['diffa_o1-b'], p)
plt.axvline(lower, ls='--', c='grey')
plt.axvline(upper, ls='--', c='grey')

plt.savefig('accdiff_1-b.png', dpi=300, bbox_inches='tight')# np.percentile(output['diffa_o3-b'], p)
plt.show()
print(f"\n{alpha*100} confidence interval {lower} and {upper}")
```
